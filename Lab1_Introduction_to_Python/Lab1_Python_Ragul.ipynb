## Problem 1: Confusion Matrix Metrics


```python
# Given values
TP = 50
FP = 8
TN = 32
FN = 10

# Total predictions
total = TP + FP + TN + FN

# Accuracy
accuracy = (TP + TN) / total

# Precision
precision = TP / (TP + FP)

# Recall
recall = TP / (TP + FN)

# F1 Score
f1_score = 2 * (precision * recall) / (precision + recall)

print('Total Predictions =', total)
print('Accuracy =', accuracy)
print('Precision =', precision)
print('Recall =', recall)
print('F1 Score =', f1_score)
```

## Problem 2: R² (Coefficient of Determination)


```python
# Given values
SSR = 420
SSE = 180

# Total Sum of Squares
SST = SSR + SSE

# R squared
R_squared = SSR / SST

print('Total Sum of Squares (SST) =', SST)
print('R² =', R_squared)
```

## Problem 3: Mean Squared Error (MSE)


```python
# Squared errors
e1 = 4
e2 = 9
e3 = 16
e4 = 25
e5 = 36

# Calculate MSE
MSE = (e1 + e2 + e3 + e4 + e5) / 5

print('Mean Squared Error (MSE) =', MSE)
```

## Problem 4: Cook’s Distance


```python
# Given values
e_squared = 9
MSE = 3
p = 3
h_ii = 0.2
n = 25

# Cook's Distance formula
D = (e_squared / (p * MSE)) * (h_ii / ((1 - h_ii) ** 2))

# Threshold
threshold = 4 / n

print('Cook’s Distance =', D)
print('Threshold (4/n) =', threshold)
print("is observationinflential :",D>threshold)
```

## Problem 5: Simple Linear Regression Prediction


```python
# Example regression equation: Y = a + bX
a = 10   # Intercept
b = 2.5   # Slope
X = 12 # Rainfall value

# Prediction
Y = a + b * X

print('Predicted Yield =', Y)
print('Intercept means yield when rainfall is 0')
print('Slope means change in yield per unit rainfall increase')
```
