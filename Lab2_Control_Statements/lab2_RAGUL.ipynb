```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = excel_df

# column names
print("Column names:")
print(df.columns)

# data types
print("\nData types:")
print(df.dtypes)

# first 5 rows
print("\nTop 5 rows:")
print(df.head())

# last 5 rows
print("\nBottom 5 rows:")
print(df.tail())

# missing values
print("\nMissing values:")
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# binary / categorical columns
categorical_cols = [
    "Gender ", "Emp", "Profile", "Cover",
    "Orientation", "Relationship", "Edu"
]

for col in categorical_cols:
    plt.figure(figsize=(5,3))
    sns.countplot(x=df[col])
    plt.title(f"frequency plot of {col}")
    plt.show()
```

4.) data visualization was performed using histograms and frequency plots to better understand the dataset. histograms were used to observe the distribution of numerical variables such as age, photos, posts, likes, and friends. frequency plots were created for categorical variables like gender, employment status, education, and relationship status to see how frequently each category appears. these visualizations help in identifying patterns, trends, and variations in the dataset and make the data easier to interpret and analyze.
