```python
import matplotlib.pyplot as plt

x = np.arange(0,3*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)


plt.plot(x,y_sin)
plt.plot(x,y_cos)
plt.xlabel("Variable x")
plt.ylabel("Trigonometric Function")
plt.title("Sin and Cos Function")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)

plt.scatter(x, y_sin, color = "blue")
plt.xlabel("Variable x")
plt.ylabel("Sin Function")
plt.title("Trigonemetric_Sin Function")

plt.subplot(2,1,2)
plt.plot(x, y_cos, color = "green")
plt.xlabel("Variable x")
plt.ylabel("Cos Function")
plt.title("Trigonemetric_Cos Function")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create variable using arange
x = np.arange(1, 10, 0.1)   # start, stop, step

# Define functions
y_linear = 2*x + 1
y_power = x**2
y_exp = np.exp(x)
y_log = np.log(x)

# 2x2 subplot
plt.figure(figsize=(10,8))

# Linear Function
plt.subplot(2,2,1)
plt.plot(x, y_linear, color='blue')
plt.title("Linear Function")
plt.grid()

# Power Function
plt.subplot(2,2,2)
plt.plot(x, y_power, color='green')
plt.title("Power Function")
plt.grid()

# Exponential Function
plt.subplot(2,2,3)
plt.plot(x, y_exp, color='red')
plt.title("Exponential Function")
plt.grid()

# Logarithmic Function
plt.subplot(2,2,4)
plt.plot(x, y_log, color='purple')
plt.title("Logarithmic Function")
plt.grid()

plt.tight_layout()
plt.show()
```
